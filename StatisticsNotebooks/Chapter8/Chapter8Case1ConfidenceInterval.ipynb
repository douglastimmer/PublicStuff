# Confidence Interval for the Mean with variance known

### by Douglas Timmer

This presentation demonstrates the calculation of a (two-sided) confidence interval for the mean with variance known. This presentation utilizes Juypter

```python
# import Python libraries and define a function
from IPython.display import display,Latex,Image, HTML
import scipy.stats as stat
import math
def roundOff(x,e):
    return math.floor(x*10**e+.5)/(10**e)
display(HTML("<h1>Step 0. Provide the sample information and value of alpha</h1>"))
#enter sample information and alpha
xbar=12.3
sigma=3.2
n=int(6)
alpha=0.01
#add print statements
#
display(HTML("<h1>Step 1. List the facts and formulas</h1>"))
display(Latex(r"""$$\mbox{Facts:}\,\bar{x}=%f,\, \sigma=%f,\,n=%d,\, \alpha=%f.$$"""%(xbar,sigma,n,alpha)))
display(HTML("<h2>Use the formula given for Case 1</h2>"))
fname1='./imgs/img077.jpg'
display(Image(filename=fname1,width=800))
#
#Convert Markdown to display
#
display(HTML("<h2>Step 2a. Using the standard normal table</h2>"))
ad2=alpha/2
oneMad2=1-ad2
zScore=stat.norm.ppf(oneMad2,0,1)
zScoreR3=roundOff(zScore,3)
answerR6=roundOff(stat.norm.cdf(zScoreR3,0,1),6)
display(Latex(r"""To find the value of $z_{%f}$, recall that we must find the upper %f quantile from the standard normal distribution. That is, one must find the value of the standard normal distribution, $z_{%f}$, such that $\Phi\left(z_{%f}\right)=1-%f=%f$. This requires looking within the standard normal tables for the value of %f and then reading the corresponding value for the $z$ random variable. This notebook uses the norm.ppf() function from SciPy (python) library to find that $z_{%f}=%f$. Rounding the value of $z_{%f}$ to two decimal places which matches the accuracy of your standard normal tables yields %f. To verify this answer is correct, examine the standard normal table and note that $\Phi\left(%f\right)=%f$. This answer is very close to the desired area of %f. Note that the Microsoft Excel function, $Norm.S.INV(%f)$ can also be used to find the $z_{%f}$."""%(ad2,ad2,ad2,ad2,ad2,oneMad2,oneMad2,ad2,zScore,ad2,zScoreR3,zScoreR3,answerR6,oneMad2,oneMad2,ad2)))
#
# convert markdown to display
#
display(HTML("<h1>Step 3. Substitute values and simplify</h1>"))
lb=xbar-zScoreR3*(sigma/math.sqrt(n))
ub=xbar+zScoreR3*(sigma/math.sqrt(n))
display(Latex(r"""$$\begin{align*}\bar{x}-z_{\alpha/2}\frac{\sigma}{\sqrt{n}}&<\mu<\bar{x}+z_{\alpha/2}\frac{\sigma}{\sqrt{n}}\\%f-z_{%f}\frac{%f}{\sqrt{%d}}&<\mu<%f+z_{%f}\frac{%f}{\sqrt{%d}}\\%f-%f\left(\frac{%f}{\sqrt{%d}}\right)&<\mu <%f+%f\left(\frac{%f}{\sqrt{%d}}\right)\\%f&<\mu<%f\end{align*}$$"""%(xbar,ad2,sigma,n,xbar,ad2,sigma,n,xbar,zScoreR3,sigma,n,xbar,zScoreR3,sigma,n,lb,ub)))
```

# Step 2. Find the value of $z_{\alpha/2}$

There are three methods to find the value of $z_{\alpha/2}$.
<ol>
<li>Use the standard normal distribution table</li>
<li>Use the t-table</li>
<li>Memorize the values</li>
</ol>

```python
display(HTML("<h2>Step 2a. Using the standard normal table</h2>"))
ad2=alpha/2
oneMad2=1-ad2
zScore=stat.norm.ppf(oneMad2,0,1)
zScoreR3=roundOff(zScore,3)
answerR6=roundOff(stat.norm.cdf(zScoreR3,0,1),6)
display(Latex(r"""To find the value of $z_{%f}$, recall that we must find the upper %f quantile from the standard normal distribution. That is, one must find the value of the standard normal distribution, $z_{%f}$, such that $\Phi\left(z_{%f}\right)=1-%f=%f$. This requires looking within the standard normal tables for the value of %f and then reading the corresponding value for the $z$ random variable. This notebook uses the norm.ppf() function from SciPy (python) library to find that $z_{%f}=%f$. Rounding the value of $z_{%f}$ to two decimal places which matches the accuracy of your standard normal tables yields %f. To verify this answer is correct, examine the standard normal table and note that $\Phi\left(%f\right)=%f$. This answer is very close to the desired area of %f. Note that the Microsoft Excel function, $Norm.S.INV(%f)$ can also be used to find the $z_{%f}$."""%(ad2,ad2,ad2,ad2,ad2,oneMad2,oneMad2,ad2,zScore,ad2,zScoreR3,zScoreR3,answerR6,oneMad2,oneMad2,ad2)))
#
display(HTML("<h2>Step 2b. Using the t-table</h2>"))
display(Latex(r"""To find the value of $z_{%f}$, a fact that will be introduced in section 2 of chapter 8 is that $z_{%f}=t_{%f,\infty}$. The upper percentiles of the t-distribution are given below. """%(ad2,ad2,ad2)))
tScore=stat.t.ppf(oneMad2,2^10000000)
tScoreR3=roundOff(tScore,3)
answerR6=roundOff(stat.norm.cdf(tScoreR3,0,1),6)
display(Latex(r"""This notebook uses the t.ppf() function from SciPy (python) library to find that $t_{%f,\infty}=%f$. Rounding the value of $t_{%f,\infty}$ to two decimal places which matches the accuracy of your standard normal tables yields %f. To verify this answer is correct, examine the standard normal table and note that $\Phi\left(%f\right)=%f$. This answer is very close to the desired area of %f. Note that the Microsoft Excel function, $T.INV(%f,\infty)$ can also be used to find the $z_{%f}$."""%(ad2,tScore,ad2,tScoreR3,tScoreR3,answerR6,oneMad2,oneMad2,ad2)))
```
