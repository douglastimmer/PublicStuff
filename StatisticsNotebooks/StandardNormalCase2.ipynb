# Standard Normal Case 2
This application calculates the probability $P(Z>b)$ for the standard normal distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sct
import math

def roundOff(x,e):
    return math.floor(x*10**e+.5)/(10**e)

b=0.2

x=np.linspace(-4,4,500)
y=sct.norm.pdf(x,0,1)
y2=0.0*x
maske=(x>b)

plt.plot(x,y,'b')
plt.fill_between(x,y,color='#666666',where=maske)
plt.plot(x,y2,'b')
plt.show()

answer=1-sct.norm.cdf(b,0,1)
print ("%8.6f" % (answer))
```
