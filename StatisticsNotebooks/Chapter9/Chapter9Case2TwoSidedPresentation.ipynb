This presentation demonstrates the the classical approach and p-value approach to conducting a two-sided hypothesis test for the mean with variance unknown. Jupyter notebooks utilize Python.

```python
#specify values for hypothesis test
from IPython.display import Image,display,Latex, HTML
import math
import scipy.stats as stat
def roundOff(x,e):
    return math.floor(x*10**e+.5)/(10**e)
#Problem 9.62
mu_0=22.5
s=0.378
n=int(5)
xbar=22.496
alpha=0.05
display(HTML("<h1>Step 0. Provide the information for the problem and the value of alpha"))
display(Latex(r"""$$\mbox{Facts:}\,\bar{x}=%f,\, s=%f,\,n=%d,\, \alpha=%f,\,\mu_0=%f.$$"""%(xbar,s,n,alpha,mu_0)))
```

```python
display(HTML("<h1>Step 1. Select the Correct Procedure</h1>"))
display(Latex(r"""Use the top line ($\mu\neq\mu_0$) of Case 2"""))
fname1='./imgs/img087Rotated.jpg'
display(Image(filename=fname1,width=800))
```

```python
display(HTML("<h2>Step 2. Define null and research hypotheses</h2>"))
display(Latex(r"""$H_0:\mu=%7.2f$"""%(mu_0)))
display(Latex(r"""$H_1:\mu\neq%7.2f$"""%(mu_0)))
```

```python
t_0=(xbar-mu_0)/(s/math.sqrt(n));
display(HTML("<h2>Step 3. Calculate the test statistic</h2>"))
display(Latex(r"""$$t_0=\frac{\bar{x}-\mu_0}{S\sqrt{n}}=\frac{%f-%f}{%f/\sqrt{%d}}=%f$$"""%(xbar,mu_0,s,n,t_0)))
```

```python
nMinus1=n-1
t_lower=stat.t.ppf(alpha/2.0,nMinus1)
t_upper=stat.t.ppf(1.0-alpha/2.0,nMinus1)
t_lower_R=roundOff(t_lower,3)
t_upper_R=roundOff(t_upper,3)
display(HTML("<h2>Step 4. Define the rejection region</h2>"))
display(Latex(r"""$\mbox{Reject }H_0\mbox{ if }t_0<-t_{\alpha/2,n-1}=-t_{%f,%d}=%f\mbox{ or }t_0>t_{\alpha/2,n-1}=t_{%f,%d}=%f$"""%(alpha/2.0,nMinus1,t_lower_R,alpha/2.0,nMinus1,t_upper_R)))
```

```python
cString=""
if t_0<t_lower or t_0>t_upper:
    cString="Since the value of the test statistic, $t_0=$%f, is in the rejection rejection region, the conclusion of the hypothesis test is to reject $H_0$. There is sufficient statistical evidence at a significance level of %f, that $\mu$ is different from %f."%(t_0,alpha,mu_0)
else:
    cString="""Since the value of the test statistic, $t_0=$%f, is not within the rejection region, the conclusion of the hypothesis test is fail to reject $H_0$. There is not sufficient statistical evidence at significance level of %f  that $\mu$ is different from %f."""%(t_0,alpha,mu_0)
display(HTML("<h2>Step 5. Evaluate the rejection region and state the conclusion</h2>"))
display(Latex(r"""%s"""%(cString)))
```

```python
p_value=2.0*(1.0-stat.t.cdf(abs(t_0),nMinus1))
display(HTML("<h2>Step 6. Find the p-value</h2>"))
display(HTML("The exact p-value for the t-test cannot found using the t-table in your textbook. Your textbook can be used to place bounds on the p-value. Python has a function to evaluate the CDF of a t-random variable."))
display(Latex(r"""$$p-value=2(1-P\left(T>|t_0|\right))=2(1-P\left(T>|%f|\right))=%f$$"""%(t_0,p_value)))
if p_value<alpha:
    cString="Since the p-value is less than alpha, we can reject the null hypothesis and conclude that the mean is different from %f at a signficance level of %f"%(mu_0,alpha)
else:
    cString="Since the p-value is NOT less than alpha, we fail to reject the null hypothesis and conclude that there is not sufficient statistical evidence that the mean is different from %f at a signficance level of %f."%(mu_0,alpha)
display(Latex(r"""%s"""%(cString)))
```
